## Wilcoxon Signed-Rank Test

The Wilcoxon Signed-Rank Test is a non-parametric test used to compare two related samples.  

In this project, we want to test whether there is a significant change between:

- respondents’ perceived effectiveness of **Green@Community** in improving Hong Kong’s recycling efficiency (before), and  
- respondents’ overall satisfaction with **Green@Community**’s effectiveness after reading the information (after).

These correspond to the two Likert-scale questions:

1. **Q_effectiveness** – “How effective do you perceive the Green@Community to be in improving Hong Kong’s recycling efficiency?”  
2. **Q_satisfaction** – “After reading the above, do you think Green@Community's overall effectiveness is satisfactory?”

Both scales range from **1** (*very ineffective/unsatisfactory*) to **4** (*very effective/satisfactory*).

### Research Question

Is there a significant difference between:

- perceived effectiveness of **Green@Community** in improving Hong Kong’s recycling efficiency, and  
- satisfaction with **Green@Community**’s overall effectiveness after reading the information?

### Null Hypothesis

\(H_0\): There is **no significant difference** between perceived effectiveness (before) and satisfaction (after).

### Alternative Hypothesis

\(H_1\): There is a **significant difference** between perceived effectiveness (before) and satisfaction (after).


```python
# ============================================================
# Wilcoxon Signed-Rank Test:
# Perceived effectiveness vs. satisfaction with Green@Community
# ============================================================

# If you haven't installed these before, uncomment and run:
# !pip install pandas scipy matplotlib openpyxl

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import wilcoxon, ttest_rel

# -------------------------
# 1. Load the dataset
# -------------------------

# Your original code
df = pd.read_excel("../GCAP3226 G@C Survey Responses_cleaned data.xlsx")

# For safety, print the column indices once (optional)
# print(list(enumerate(df.columns)))

# We will select the two columns by **position** to avoid issues with long text headers:
# Column 6  -> "How effective do you perceive the Green@Community to be in improving Hong Kong’s recycling efficiency?"
# Column 17 -> "After reading the above, do you think Green@Community's overall effectiveness is satisfactory?"
col_effectiveness = df.columns[6]
col_satisfaction = df.columns[17]

# Extract the two series
effectiveness_raw = df[col_effectiveness]
satisfaction_raw = df[col_satisfaction]

# -------------------------
# 2. Clean and align the data
# -------------------------

# Make sure we only keep rows where BOTH answers are not missing
common_indices = df.dropna(subset=[col_effectiveness, col_satisfaction]).index
effectiveness = df.loc[common_indices, col_effectiveness].astype(float)
satisfaction = df.loc[common_indices, col_satisfaction].astype(float)

print("Data Summary:")
print("======================================")
print(f"Number of paired observations: {len(effectiveness)}\n")

print("Perceived effectiveness (before):")
print(effectiveness.describe())
print("\nSatisfaction after reading info (after):")
print(satisfaction.describe())

# -------------------------
# 3. Calculate the difference (After - Before)
# -------------------------
difference = satisfaction - effectiveness

print("\nDifference (After - Before):")
print(difference.describe())
print(f"\nNumber of positive changes (After > Before): {(difference > 0).sum()}")
print(f"Number of negative changes (After < Before): {(difference < 0).sum()}")
print(f"Number of no changes      (After = Before): {(difference == 0).sum()}")

# -------------------------
# 4. Wilcoxon Signed-Rank Test
# -------------------------
# This test compares the median of the paired differences to zero
statistic, p_value = wilcoxon(effectiveness, satisfaction, alternative='two-sided')

print("\nWilcoxon Signed-Rank Test Results:")
print("======================================")
print(f"Test Statistic: {statistic:.4f}")
print(f"P-value:        {p_value:.6f}")
print("Significance Level: α = 0.05")

# -------------------------
# 5. Interpretation
# -------------------------
print("\nInterpretation:")
if p_value < 0.05:
    print("✓ REJECT the null hypothesis (H₀)")
    print("✓ There IS a statistically significant difference")
    print("  between perceived effectiveness and satisfaction after reading the information.")
else:
    print("✗ FAIL to reject the null hypothesis (H₀)")
    print("✗ There is NO statistically significant difference")
    print("  between perceived effectiveness and satisfaction after reading the information.")

# Direction of change
mean_diff = difference.mean()
median_diff = difference.median()
print(f"\nMean change (After - Before): {mean_diff:.3f}")
print(f"Median change (After - Before): {median_diff:.3f}")

if mean_diff > 0:
    print("On average, satisfaction scores are HIGHER than initial perceived effectiveness.")
elif mean_diff < 0:
    print("On average, satisfaction scores are LOWER than initial perceived effectiveness.")
else:
    print("On average, there is no change between the two scores.")

# -------------------------
# 6. Effect Size (Cohen's d approximation)
# -------------------------
# We approximate effect size by converting the paired t-test statistic to Cohen's d
t_stat, _ = ttest_rel(satisfaction, effectiveness)
n = len(effectiveness)
cohens_d = t_stat / np.sqrt(n)

print("\nEffect Size (Cohen's d approximation):")
print(f"Cohen's d: {cohens_d:.4f}")

if abs(cohens_d) < 0.2:
    effect_size = "negligible"
elif abs(cohens_d) < 0.5:
    effect_size = "small"
elif abs(cohens_d) < 0.8:
    effect_size = "medium"
else:
    effect_size = "large"

print(f"Effect size interpretation: {effect_size}")

# -------------------------
# 7. Visualisations
# -------------------------
plt.style.use("default")

fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# 7.1 Histogram of differences
axes[0].hist(difference, bins=15, alpha=0.7, edgecolor='black')
# <<< changed: mean line color to red >>>
axes[0].axvline(difference.mean(), color='red', linestyle='--',
                label=f'Mean: {difference.mean():.2f}')
axes[0].set_title('Distribution of Changes\n(After - Before)')
axes[0].set_xlabel('Change in Score (Satisfaction - Effectiveness)')
axes[0].set_ylabel('Frequency')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# 7.2 Bar chart of change categories
change_categories = ['Decreased', 'No Change', 'Increased']
change_counts = [
    (difference < 0).sum(),
    (difference == 0).sum(),
    (difference > 0).sum()
]

# <<< changed: 3 different colours for the bars >>>
bar_colors = ['red', 'gray', 'green']  # decreased, no change, increased
bars = axes[1].bar(change_categories, change_counts,
                   alpha=0.7, color=bar_colors)

axes[1].set_title('Number of Respondents by Change Category')
axes[1].set_ylabel('Number of Respondents')
axes[1].grid(True, alpha=0.3)

# Add value labels on bars
for bar, count in zip(bars, change_counts):
    height = bar.get_height()
    axes[1].text(bar.get_x() + bar.get_width() / 2., height + 0.5,
                 f'{count}', ha='center', va='bottom')

plt.tight_layout()
plt.show()

# -------------------------
# 8. Summary of Changes
# -------------------------
print("\nSummary of Changes:")
print(f"Mean change: {difference.mean():.3f}")
print(f"Median change: {difference.median():.3f}")
print(f"Standard deviation of changes: {difference.std():.3f}")
print(f"Range of changes: {difference.min():.1f} to {difference.max():.1f}")
```
